# FCFS

```python
import pandas as pd
```

```python

def FCFS(n,process,burst_time,arrival_time):
  df = pd.DataFrame(columns = ['process','burst_time','arrival_time','completion_time','turnaround_time','waiting_time'])
  for i in range(n):
    df['process'] = process
    df['burst_time']= burst_time
    df['arrival_time'] = arrival_time

  df = df.sort_values('arrival_time')
  sum =0
  for i in range(n):
    sum += burst_time[i]
    df['completion_time'][i] = sum

  df['turnaround_time'] = df['completion_time'] - df['arrival_time']
  df['waiting_time'] = df['turnaround_time'] - df['burst_time']

  print('Avg waiting time:',df['waiting_time'].mean())
  print('\n\n')

  print(df)
```

```python
p = ['p1','p2','p3']
bt = [ 24, 3, 4]
at = [0,0,0]
n = 3
FCFS(n,p,bt,at)
```

```python
def SFJ(n,process,burst_time,arrival_time,priority):
  df1 = pd.DataFrame(columns = ['process','burst_time','arrival_time','completion_time','turnaround_time','waiting_time','priority'])
  for i in range(0,n):
    df1['process'] = process
    df1['burst_time']= burst_time
    df1['arrival_time'] = arrival_time
    #df1['priority'] = priority

  df1 = df1.sort_values('burst_time')
  df1=df1.reset_index()
  sum =0
  for i in range(n):
    sum =sum+ df1['burst_time'][i]
    df1['completion_time'][i] = sum

  df1['turnaround_time'] = df1['completion_time'] - df1['arrival_time']
  df1['waiting_time'] = df1['turnaround_time'] - df1['burst_time']

  print('Avg waiting time:',df1['waiting_time'].mean())

  print(df1)
```

# Priority

```python
def Priority(n,process,burst_time,arrival_time,priority):
  df2 = pd.DataFrame(columns = ['process','burst_time','arrival_time','completion_time','turnaround_time','waiting_time','priority'])
  for i in range(0,n):
    df2['process'] = process
    df2['burst_time']= burst_time
    df2['arrival_time'] = arrival_time
    df2['priority'] = priority

  df2 = df2.sort_values('priority')
  sum =0
  for i in range(0,n):
    sum += df2['burst_time'][i]
    df2['completion_time'][i] = sum

  df2['turnaround_time'] = df2['completion_time'] - df2['arrival_time']
  df2['waiting_time'] = df2['turnaround_time'] - df2['burst_time']

  print('Avg waiting time:',df2['waiting_time'].mean())

  print(df2)
```

# FIFO

```python
def FIFO (pages,frames):
  page_faults = 0
  page_frames = [None] * frames
  frame_pointer = 0

  for page in pages :
    if page not in page_frames :
      page_frames[frame_pointer] = page
      frame_pointer = (frame_pointer + 1) % frames
      page_faults+=1
    print(page_frames)

  hit_ratio = 1 - (page_faults/len(pages))
  return page_faults , hit_ratio
```

```python
pages = [1,3,0,3,5,6]
frames = 3

faults,hit_ratio = FIFO(pages ,frames)
print("page_faults are : ", faults)
print("hit ratio is: ", hit_ratio)
```

# LRU

```python
def LRU(pages,frames) :
  page_faults = 0
  page_frames = [None] * frames
  page_usage = [-1] * frames

  for i , page in enumerate (pages):
    if page not in page_frames :
      if None in page_frames :
        page_frames[page_frames.index(None)] = page
      else:
        page_frames[page_usage.index(min(page_usage))] = page
      page_faults+=1
    page_usage[page_frames.index(page)] = i
    print(page_frames)

  hit_ratio = 1 - (page_faults/len(pages))
  return page_faults , hit_ratio
```

```python
pages = [7, 0, 1, 2, 0, 3, 0, 4, 2, 3, 0, 3, 2]
frames = 4

faults,hit_ratio = LRU(pages ,frames)
print("page_faults are : ", faults)
print("hit ratio is: ", hit_ratio)
```

# OPTIMAL

```python
def OPTIMAL(pages ,frames):
  page_faults = 0
  page_frames = [None] * frames

  for i,page in enumerate (pages):
    if page not in page_frames:
      if None in page_frames:
        page_frames[page_frames.index(None)] = page
      else:
        replace_page = max(page_frames, key=lambda x: pages[i+1:].index(x) if x in pages[i+1:] else float('inf'))
        page_frames[page_frames.index(replace_page)] = page
      page_faults+=1
    print(page_frames)
  
  hit_ratio = 1 - (page_faults / len(pages))
  return page_faults, hit_ratio
```

```python
pages = [7, 0, 1, 2, 0, 3, 0, 4, 2, 3, 0, 3, 2]
frames = 4

faults, hit_ratio = OPTIMAL(pages, frames)
print("Number of Page Faults using Optimal:", faults)
print("Hit Ratio using Optimal:", hit_ratio)
```

# FIRST FIT

```python
def FIRST(blocksize , m , processsize, n):

  allocation = [-1] * n
  for i in range(n):
    for j in range(m):
      if (blocksize[j]>=processsize[i]):
        allocation[i]=j
        blocksize[j]-=processsize[i]
        break
  
  print("Process No  Process size  Blocksize")
  for i in range(n):
    print("   ",i+1, "   ",processsize[i],"   ",end=" ")

    if allocation[i]!=-1:
      print(allocation[i]+1)
    else:
      print("Not Allocated")
```

```python
blockSize = [100, 500, 200, 450, 600]
processSize = [212, 417, 112, 426]
m = len(blockSize)
n = len(processSize)

FIRST(blockSize,m,processSize,n)
```

# BEST FIT

```python
def BEST(blocksize , m , processsize, n):

  allocation = [-1] * n
  for i in range(n):
    idx=-1
    for j in range(m):
      if (blocksize[j]>=processsize[i]):
        if idx == -1:
          idx=j
        elif blocksize[idx] > blocksize[j]:
          idx=j

    if idx!=-1:
      allocation[i]=idx
      blocksize[idx]-=processsize[i]
  
  print("Process No  Process size  Blocksize")
  for i in range(n):
    print("   ",i+1, "   ",processsize[i],"   ",end=" ")

    if allocation[i]!=-1:
      print(allocation[i]+1)
    else:
      print("Not Allocated")
```

```python
blockSize = [100, 500, 200, 450, 600]
processSize = [212, 417, 112, 426]
m = len(blockSize)
n = len(processSize)

BEST(blockSize,m,processSize,n)
```

# SSTF

```python
def calculateDifference(queue, head, diff):
	for i in range(len(diff)):
		diff[i][0] = abs(queue[i] - head)
	
# find unaccessed track which is
# at minimum distance from head
def findMin(diff):

	index = -1
	minimum = 999999999

	for i in range(len(diff)):
		if (not diff[i][1] and
				minimum > diff[i][0]):
			minimum = diff[i][0]
			index = i
	return index
	
def SSTF(request, head):			
		if (len(request) == 0):
			return
		
		l = len(request)
		diff = [0] * l
		
		# initialize array
		for i in range(l):
			diff[i] = [0, 0]
		
		# count total number of seek operation	
		seek_count = 0
		
		# stores sequence in which disk
		# access is done
		seek_sequence = [0] * (l + 1)
		
		for i in range(l):
			seek_sequence[i] = head
			calculateDifference(request, head, diff)
			index = findMin(diff)
			
			diff[index][1] = True
			
			# increase the total count
			seek_count += diff[index][0]
			
			# accessed track is now new head
			head = request[index]
	
		# for last accessed track
		seek_sequence[len(seek_sequence) - 1] = head
		
		print("Total number of seek operations =",
									seek_count)
														
		print("Seek Sequence is")
		
		# print the sequence
		for i in range(l + 1):
			print(seek_sequence[i])
```

```python
arr = [ 82,170,43,140,24,16,190]
head = 50

SSTF(arr, head)
```

# SCAN

```python
size = 7
disk_size = 200

def SCAN(arr, head, direction):

	seek_count = 0
	distance, cur_track = 0, 0
	left = []
	right = []
	seek_sequence = []

	if (direction == "left"):
		left.append(0)
	elif (direction == "right"):
		right.append(disk_size - 1)

	for i in range(size):
		if (arr[i] < head):
			left.append(arr[i])
		if (arr[i] > head):
			right.append(arr[i])

	left.sort()
	right.sort()

	run = 2
	while (run != 0):
		if (direction == "left"):
			for i in range(len(left) - 1, -1, -1):
				cur_track = left[i]

				# Appending current track to
				# seek sequence
				seek_sequence.append(cur_track)

				# Calculate absolute distance
				distance = abs(cur_track - head)

				# Increase the total count
				seek_count += distance

				# Accessed track is now the new head
				head = cur_track
			
			direction = "right"
	
		elif (direction == "right"):
			for i in range(len(right)):
				cur_track = right[i]
				
				# Appending current track to seek
				# sequence
				seek_sequence.append(cur_track)

				# Calculate absolute distance
				distance = abs(cur_track - head)

				# Increase the total count
				seek_count += distance

				# Accessed track is now new head
				head = cur_track
			
			direction = "left"
		
		run -= 1

	print("Total number of seek operations =",
		seek_count)

	print("Seek Sequence is")

	for i in range(len(seek_sequence)):
		print(seek_sequence[i])
```

```python
arr = [ 82,170,43,140,24,16,190]
head = 50
direction = "right"

SCAN(arr, head, direction)

```

# C-SCAN

```python
size = 7
disk_size = 200


def CSCAN(arr, head):

	seek_count = 0
	distance = 0
	cur_track = 0
	left = []
	right = []
	seek_sequence = []

	left.append(0)
	right.append(disk_size - 1)

	for i in range(size):
		if (arr[i] < head):
			left.append(arr[i])
		if (arr[i] > head):
			right.append(arr[i])

	left.sort()
	right.sort()

	for i in range(len(right)):
		cur_track = right[i]

		# Appending current track
		# to seek sequence
		seek_sequence.append(cur_track)

		# Calculate absolute distance
		distance = abs(cur_track - head)

		# Increase the total count
		seek_count += distance

		# Accessed track is now new head
		head = cur_track

	# Once reached the right end
	# jump to the beginning.
	head = 0

	# adding seek count for head returning from 199 to 0
	seek_count += (disk_size - 1)

	for i in range(len(left)):
		cur_track = left[i]

		# Appending current track
		# to seek sequence
		seek_sequence.append(cur_track)

		# Calculate absolute distance
		distance = abs(cur_track - head)

		# Increase the total count
		seek_count += distance

		# Accessed track is now the new head
		head = cur_track

	print("Total number of seek operations =",seek_count)
	print("Seek Sequence is")
	print(*seek_sequence, sep="\n")

```

```python
arr = [ 82,170,43,140,24,16,190]
head = 50
direction = "right"

CSCAN(arr, head)
```
